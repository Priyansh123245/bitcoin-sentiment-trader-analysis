```python
import gdown

# Historical trader data
url1 = "https://drive.google.com/uc?id=1IAfLZwu6rJzyWKgBToqwSmmVYU6VbjVs"
gdown.download(url1, "historical_data.csv", quiet=False)

# Fear Greed Index
url2 = "https://drive.google.com/uc?id=1PgQC0tO8XN-wqkNyghWc_-mnrYv_nhSf"
gdown.download(url2, "fear_greed_index.csv", quiet=False)
```

```python
trader_df = pd.read_csv("historical_data.csv")
sentiment_df = pd.read_csv("fear_greed_index.csv")

print(trader_df.shape, sentiment_df.shape)
trader_df.head()
```

```python
print(trader_df.info())
print("-"*50)
print(sentiment_df.info())
```

```python
print(trader_df.isnull().sum())
print("-"*50)
print(sentiment_df.isnull().sum())
```

```python
# Trader data ka time column datetime me convert karo
# Column ka naam file ke hisaab se check kar lena (time / Time / timestamp)
time_col = [c for c in trader_df.columns if 'time' in c.lower()][0]
trader_df['datetime'] = pd.to_datetime(trader_df[time_col], errors='coerce')
trader_df['date'] = trader_df['datetime'].dt.date

# Sentiment data ka Date column
date_col = [c for c in sentiment_df.columns if 'date' in c.lower()][0]
sentiment_df['date'] = pd.to_datetime(sentiment_df[date_col], errors='coerce').dt.date

trader_df.dropna(subset=['date'], inplace=True)
sentiment_df.dropna(subset=['date'], inplace=True)
```

```python
class_col = [c for c in sentiment_df.columns if 'class' in c.lower()][0]
sentiment_small = sentiment_df[['date', class_col]].rename(columns={class_col: 'sentiment'})

merged = pd.merge(trader_df, sentiment_small, on='date', how='left')
print(merged.shape)
merged.head()
```

```python
pnl_col = [c for c in merged.columns if 'pnl' in c.lower()][0]
merged[pnl_col] = pd.to_numeric(merged[pnl_col], errors='coerce')
merged.dropna(subset=[pnl_col], inplace=True)
print(pnl_col)
```

```python
avg_pnl = merged.groupby('sentiment')[pnl_col].mean().sort_values(ascending=False)
print(avg_pnl)

plt.figure(figsize=(8,5))
avg_pnl.plot(kind='bar', color='teal')
plt.title("Average Closed PnL by Market Sentiment")
plt.ylabel("Avg Closed PnL")
plt.xticks(rotation=45)
plt.show()
```

Average closed PnL is highest during Greed and lowest during Fear, showing that traders tend to perform better when market sentiment is optimistic.

```python
merged['is_win'] = merged[pnl_col] > 0
win_rate = merged.groupby('sentiment')['is_win'].mean() * 100
print(win_rate)

plt.figure(figsize=(8,5))
win_rate.sort_values(ascending=False).plot(kind='bar', color='orange')
plt.title("Win Rate (%) by Market Sentiment")
plt.ylabel("Win Rate %")
plt.xticks(rotation=45)
plt.show()
```

Win rate is higher during Extreme Greed, but average profit per trade is lower, suggesting traders book small profits quickly instead of holding for bigger gains.

```python
lev_col = [c for c in merged.columns if 'lever' in c.lower()]
if lev_col:
    lev_col = lev_col[0]
    merged[lev_col] = pd.to_numeric(merged[lev_col], errors='coerce')
    avg_lev = merged.groupby('sentiment')[lev_col].mean()
    print(avg_lev)

    plt.figure(figsize=(8,5))
    avg_lev.sort_values(ascending=False).plot(kind='bar', color='purple')
    plt.title("Average Leverage by Market Sentiment")
    plt.ylabel("Avg Leverage")
    plt.xticks(rotation=45)
    plt.show()
```

Traders use higher leverage during Fear and Extreme Fear, indicating riskier behavior when the market is uncertain.

```python
side_col = [c for c in merged.columns if 'side' in c.lower()][0]
side_sentiment = merged.groupby(['sentiment', side_col]).size().unstack(fill_value=0)
side_pct = side_sentiment.div(side_sentiment.sum(axis=1), axis=0) * 100
print(side_pct)

side_pct.plot(kind='bar', stacked=True, figsize=(9,5), colormap='coolwarm')
plt.title("Long vs Short Distribution by Sentiment")
plt.ylabel("Percentage")
plt.xticks(rotation=45)
plt.show()
```

Long positions dominate during Greed while short positions increase during Fear, showing that traders closely follow market sentiment.

```python
size_col = [c for c in merged.columns if 'size' in c.lower()][0]
merged[size_col] = pd.to_numeric(merged[size_col], errors='coerce')
avg_size = merged.groupby('sentiment')[size_col].mean()
print(avg_size)

plt.figure(figsize=(8,5))
avg_size.sort_values(ascending=False).plot(kind='bar', color='green')
plt.title("Average Trade Size by Sentiment")
plt.xticks(rotation=45)
plt.show()
```

Average trade size is larger during Greed, reflecting higher confidence, while trade sizes shrink during Fear due to cautious behavior.

```python
daily = merged.groupby('date').agg(daily_pnl=(pnl_col,'sum')).reset_index()
daily = daily.merge(sentiment_small, on='date', how='left')
daily['date'] = pd.to_datetime(daily['date'])
daily.sort_values('date', inplace=True)

plt.figure(figsize=(14,6))
plt.plot(daily['date'], daily['daily_pnl'], color='black', linewidth=1)
plt.title("Daily Aggregate PnL Over Time")
plt.xlabel("Date")
plt.ylabel("Total PnL")
plt.show()
```

Sudden spikes in daily PnL often occur around dates where sentiment shifts sharply, showing that changing sentiment creates short-term volatility and trading opportunities.

```python
num_cols = [pnl_col, size_col]
if lev_col: num_cols.append(lev_col)

corr = merged[num_cols].corr()
plt.figure(figsize=(6,5))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

Leverage shows a weak correlation with PnL, meaning higher leverage does not guarantee higher profit and mainly increases risk.

```python

print(f"1. Highest avg PnL sentiment: {avg_pnl.idxmax()} ({avg_pnl.max():.2f})")
print(f"2. Lowest avg PnL sentiment: {avg_pnl.idxmin()} ({avg_pnl.min():.2f})")
print(f"3. Highest win rate sentiment: {win_rate.idxmax()} ({win_rate.max():.2f}%)")
if lev_col:
    print(f"4. Highest avg leverage sentiment: {avg_lev.idxmax()} ({avg_lev.max():.2f})")
print(f"5. Total trades analyzed: {len(merged)}")
```

Overall, trader performance depends not just on market sentiment but also on trader behavior like leverage usage, position sizing, and timing, making sentiment a useful signal rather than a standalone strategy.
